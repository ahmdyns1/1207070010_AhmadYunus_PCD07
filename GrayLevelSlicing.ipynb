```python
#Import Library
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# membaca gambar di folder
image = cv2.imread('gambar1.jpg', 0)
```

```python
# mendapatkan jumlah baris dan kolom gambar
row, column = image.shape

# membuat array baru 
image1 = np.zeros((row, column),dtype = 'uint8')

# menghitung nilai rentang terkecil dan terbesar
min_range = 10
max_range = 60

# melakukan iterasi setiap piksel menggunakan dua loop
for i in range(row):
    for j in range(column):
        if image[i,j]>min_range and image[i,j]<max_range:
            image1[i,j] = 255 # mengatur piksel menjadi putih (255) jika nilainya berada dalam range
        else:
            image1[i,j] = 0 # mengatur piksel menjadi hitam (0) jika nilainya di luar range
```

```python
# membuat sebuah gambar dengan ukuran 12x12 dan axes 2x2 
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

#menampilkan citra beserta title dan histogramnya
ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(image.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(image1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(image1.ravel(), bins=256)
ax[3].set_title('Histogram Output')

fig.tight_layout() 
```
